### Partie 2 : Liens et corrélations entre variables  

### 2.1. Convertir l'âge en une valeur numérique

```python
# On crée une copy de data afin de ne pas altérer le fichier original
data2 = data.copy()
```

```python
# On crée une fonction qui prend en paramètre d'entrée une chaine de caractères
def conversion(agestr): 
    
    # Si la chaine de caractère est nulle alors on remplace la valeur de la ligne par None
    if pd.isnull(agestr): 
        return None
    
    # On crée une liste qui contiendra les différents éléments du split de la colonne "Age"
    liste = agestr.split()
    
    # On récupère le premier élément du split
    agestr = liste[0]
    
    # Si ce premier élément contient un + comme 60+, celui-ci est remplacé par une chaine de caractère vide
    agestr = agestr.replace("+", "")
    
    # On convertit la chaine de caractères en integer afin de pouvoir faire des statistiques plus tard
    agestr = int(agestr)
    
    return agestr
```

```python
# On utilise la méthode .apply() afin d'appliquer la fonction conversion sur l'ensemble des lignes de la colonne "Age"
# On assigne le résultat à une nouvelle colonne int_age
data2["int_age"] = data2["Age"].apply(conversion)
```

```python
# On applique la méthode .describe() afin d'obtenir des statistiques descriptives sur la colonne "int_age"
data2["int_age"].describe()
```

Conclusion:
- On fait un décompte de l'âge sur 947 personnes dont la moyenne d'âge est de 40,1 ans avec un écart-type de 15,4ans. 
- Le plus jeune participant à 18ans et le plus vieux 60ans. 
- 50% des participants ont 45ans et moins. 

### 2.2 Convertir les revenus en valeurs numériques

```python
# On crée une fonction qui prend en paramètre d'entrée une chaine de caractères
def money_convert(moneystr):
    
    # Si la chaine de caractère est nulle alors on remplace la valeur de la ligne par None
    if pd.isnull(moneystr): 
        return None
    
    # On crée une liste qui contiendra les différents éléments du split de la colonne répertoriant les revenus annuels communs
    liste = moneystr.split()
    
    # On récupère le premier élément du split 
    moneystr = liste[0]
    
    # Si ce premier élément est "Prefer" on retourne None
    if moneystr == "Prefer": 
        return None
    
    # On remplace les signes $ ou , par une chaine de caractères vide
    # On utilise les expressions régulières 
    moneystr = moneystr.replace('$',"").replace(',',"")
   
    # On convertit la chaine de caractères en integer afin de pouvoir faire des statistiques plus tard
    moneystr = int(moneystr)
    
    return moneystr
```

```python
# On utilise la méthode .apply() afin d'appliquer la fonction money_convert sur l'ensemble des lignes de la colonne d'intérêt
# On assigne le résultat à une nouvelle colonne int_income
# Par défaut la fonction apply s'applique sur la colonne
data2["int_income"] = data2["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(money_convert)
```

```python
# On applique la méthode .describe() afin d'obtenir des statistiques descriptives sur la colonne "int_income"
data2["int_income"].describe()
```

Conclusion:
- On fait un décompte du revenu sur 829 personnes dont le revenu moyen est de 75 965 dollars avec un écart-type de 59 069 dollars. 
- Le revenu le plus faible est de 0 dollars et le plus élevé est de 200 000 dollars. 
- 50% des participants ont des revenus de 75 000 dollars et moins. 

### 2.3 Corrélation entre distance et revenus

```python
# On crée un masque qui retournera un booléen True si les valeurs dans la colonne int_income sont inférieures à 150000
masque_inf_150K = data2["int_income"] < 150000

# On applique le masque à l'ensemble du jeu de données pour récupérer les valeurs adéquates
salaire_bas = data2[masque_inf_150K]
```

```python
# On utilise la méthode .value_counts() pour connaitre les habitudes de voyages pour Thanksgiving
# Ainsi, on sait que parmi les familles gagnant moins de 150K, 281 ne voyagent pas du tout à Thanksgiving.
salaire_bas["How far will you travel for Thanksgiving?"].value_counts()
```

Conclusion : Parmi les personnes avec moins de 150 000 dollars de revenus, 281 ne voyagent pas du tout à Thanksgiving tandis que 55 voyagent en dehors de la ville.

```python
# On applique la même méthode que précédemment pour les familles gagnant plus de 150K
masque_sup_150K = data2["int_income"] >= 150000
salaire_haut = data2[masque_sup_150K]
```

```python
# Ainsi, on sait que parmi les familles gagnant plus de 150K, 66 ne voyagent pas du tout à Thanksgiving.
salaire_haut["How far will you travel for Thanksgiving?"].value_counts()
```

Conclusion : Parmi les personnes avec plus de 150 000 dollars de revenus, 66 ne voyagent pas du tout à Thanksgiving tandis que 15 voyagent en dehors de la ville.
