### Partie 2 : Liens et corrÃ©lations entre variables  

### 2.1. Convertir l'Ã¢ge en une valeur numÃ©rique

```python
# On crÃ©e une copy de data afin de ne pas altÃ©rer le fichier original
data2 = data.copy()
```

```python
# On crÃ©e une fonction qui prend en paramÃ¨tre d'entrÃ©e une chaine de caractÃ¨res
def conversion(agestr): 
    
    # Si la chaine de caractÃ¨re est nulle alors on remplace la valeur de la ligne par None
    if pd.isnull(agestr): 
        return None
    
    # On crÃ©e une liste qui contiendra les diffÃ©rents Ã©lÃ©ments du split de la colonne "Age"
    liste = agestr.split()
    
    # On rÃ©cupÃ¨re le premier Ã©lÃ©ment du split
    agestr = liste[0]
    
    # Si ce premier Ã©lÃ©ment contient un + comme 60+, celui-ci est remplacÃ© par une chaine de caractÃ¨re vide
    agestr = agestr.replace("+", "")
    
    # On convertit la chaine de caractÃ¨res en integer afin de pouvoir faire des statistiques plus tard
    agestr = int(agestr)
    
    return agestr
```

```python
# On utilise la mÃ©thode .apply() afin d'appliquer la fonction conversion sur l'ensemble des lignes de la colonne "Age"
# On assigne le rÃ©sultat Ã  une nouvelle colonne int_age
data2["int_age"] = data2["Age"].apply(conversion)
```

```python
# On applique la mÃ©thode .describe() afin d'obtenir des statistiques descriptives sur la colonne "int_age"
data2["int_age"].describe()
```

Conclusion:
- On fait un dÃ©compte de l'Ã¢ge sur 947 personnes dont la moyenne d'Ã¢ge est de 40,1 ans avec un Ã©cart-type de 15,4ans. 
- Le plus jeune participant Ã  18ans et le plus vieux 60ans. 
- 50% des participants ont 45ans et moins. 

### 2.2 Convertir les revenus en valeurs numÃ©riques

```python
# On crÃ©e une fonction qui prend en paramÃ¨tre d'entrÃ©e une chaine de caractÃ¨res
def money_convert(moneystr):
    
    # Si la chaine de caractÃ¨re est nulle alors on remplace la valeur de la ligne par None
    if pd.isnull(moneystr): 
        return None
    
    # On crÃ©e une liste qui contiendra les diffÃ©rents Ã©lÃ©ments du split de la colonne rÃ©pertoriant les revenus annuels communs
    liste = moneystr.split()
    
    # On rÃ©cupÃ¨re le premier Ã©lÃ©ment du split 
    moneystr = liste[0]
    
    # Si ce premier Ã©lÃ©ment est "Prefer" on retourne None
    if moneystr == "Prefer": 
        return None
    
    # On remplace les signes $ ou , par une chaine de caractÃ¨res vide
    # On utilise les expressions rÃ©guliÃ¨res 
    moneystr = moneystr.replace('$',"").replace(',',"")
   
    # On convertit la chaine de caractÃ¨res en integer afin de pouvoir faire des statistiques plus tard
    moneystr = int(moneystr)
    
    return moneystr
```

```python
# On utilise la mÃ©thode .apply() afin d'appliquer la fonction money_convert sur l'ensemble des lignes de la colonne d'intÃ©rÃªt
# On assigne le rÃ©sultat Ã  une nouvelle colonne int_income
# Par dÃ©faut la fonction apply s'applique sur la colonne
data2["int_income"] = data2["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(money_convert)
```

```python
# On applique la mÃ©thode .describe() afin d'obtenir des statistiques descriptives sur la colonne "int_income"
data2["int_income"].describe()
```

Conclusion:
- On fait un dÃ©compte du revenu sur 829 personnes dont le revenu moyen est de 75 965 dollars avec un Ã©cart-type de 59 069 dollars. 
- Le revenu le plus faible est de 0 dollars et le plus Ã©levÃ© est de 200 000 dollars. 
- 50% des participants ont des revenus de 75 000 dollars et moins.
